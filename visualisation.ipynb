```python
import matplotlib.pyplot as plt
import pandas_datareader as pdr
from technical_indicators import *
```

```python
start = "2019-01-01"
end = "2020-12-31"
price_data = pdr.DataReader("^FCHI", "yahoo", start=start, end=end)
```

```python
price_data["SMA 20"] = sma(price_data["Adj Close"], 20)
price_data["EMA 40"] = ema(price_data["Adj Close"], 40)
price_data["EMA 10"] = ema(price_data["Adj Close"], 10)
```

```python
fig, ax = plt.subplots()

fig.set_size_inches(20, 10)

price_data["Adj Close"].plot(ax=ax)
#price_data["SMA 20"].plot(ax=ax)
price_data["EMA 40"].plot(ax=ax)
price_data["EMA 10"].plot(ax=ax)
```
